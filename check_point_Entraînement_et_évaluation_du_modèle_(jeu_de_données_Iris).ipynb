```python
##import bibliothèque
##remplacer les valeurs manquantes
##charger le dataset directement depuis sklearn
##normaliser les données
##séparer les données en deux parties
##modèle arbre de decision
##modèle randomforest
##evaluation du modèle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score
```

```python
##chargé datset iris dans la variable donne et séparer lees données et les reponses
donne = load_iris()
X = donne.data

y = donne.target
```

```python
##transformer les données brute en tableau claire
X = pd.DataFrame(X, columns = donne.feature_names)
X.head()
```

```python
##vérification si valeurs manquantes
X.isna().sum().sum()
```

```python
##affiché resumé complet dataframe X
X.info()
```

```python
##afficher statistiques descriptive ou analyse exploratoire
X.describe()
```

```python
##transformer les tables en tableau lisibles pour mieux visualiser
y = pd.DataFrame(y, columns = ["target"])
y.head()
```

```python
##compter le nombre d'exemplaire dans chaque espèces de fleurs pour si données equilibrées
y.value_counts()
```

```python
##diviser le dataset en deux parties: entreinement et test
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = .2,
                                                 stratify = y)
```

```python
##compter nombre d'exemplaire de chaque espèce dans données a entrainer si equilibrées
y_train.value_counts()
```

Maquette de bâtiment

```python
##créer et entreiner le modèle d'arbre de décision pour classification de fleurs
model_dt = DecisionTreeClassifier()
model_dt.fit(X_train,y_train)
```

Prédiction et évaluation

```python
##test modèle entrainer et mesure de ces erreurs de classification
y_pred_dt = model_dt.predict(X_test)
confusion_matrix(y_test,y_pred_dt)
```

```python
##évaluation detaillée de la performance du modèle
print(classification_report(y_test,y_pred_dt))
```

```python
##créer et entrainer un modèle Random Forest plus robuste et puissante
model_rf = RandomForestClassifier(n_estimators= 200, max_depth=8)
model_rf.fit(X_train,y_train)
```

```python
##test et evaluation du modèle Random forest
y_pred_rf = model_rf.predict(X_test)

print(confusion_matrix(y_test,y_pred_rf))

print(classification_report(y_test,y_pred_rf))
```

```python
##optimisation automatique de modèle Random Forest en testant plusieur combinaisons de parametres pour trouver la meilleure
from sklearn.model_selection import GridSearchCV

param_grid = {
    "n_estimators": [100, 200,300,350,400],
    "max_depth": [5, 8,10,None]
}

grid = GridSearchCV(model_rf, param_grid, cv=5, scoring="roc_auc")
grid.fit(X_train, y_train)

print("Best params:", grid.best_params_)
```

```python
##afficher le meilleur parametre automatique trouvé par gridsearchcv
print("Best params:", grid.best_params_)
```

```python
##créer et entrainer le modèle Random Forest avec des parametres fixes
model_rf = RandomForestClassifier(n_estimators= 100, max_depth=5)
model_rf.fit(X_train,y_train)
```

```python
##évalauation des performances du modèle random forest entrainé
y_pred_rf_best = model_rf.predict(X_test)

print(confusion_matrix(y_test,y_pred_rf_best))

print(classification_report(y_test,y_pred_rf_best))
```
